```python
url = "https://api.aerisapi.com/fires/"
```

```python
clientID = "&client_id=[]&client_secret=[]" #replace brackets with id strings
formatSTR = "&format=geojson"
query = "search?query=state:ca,area!:NULL,area!:None" #replace ca with respective state abbr.
limit = "&limit=250"
```

```python
queryURL = url + query + formatSTR + limit + clientID
print(queryURL)
```

```python
fire_response = requests.get(queryURL)
fireJSON = fire_response.json()

pprint(fireJSON)
```
